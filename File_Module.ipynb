```python
from docx import Document

def is_highlighted(run):
    return run.font.highlight_color is not None

def process_question(paragraph):
    options = []
    correct_option = None

    for run in paragraph.runs:
        text = run.text.strip()
        options.append(text)

        if is_highlighted(run):
            correct_option = text

    return options, correct_option

def process_input_file(input_file):
    data = []

    doc = Document(input_file)

    for paragraph in doc.paragraphs:
        options, correct_option = process_question(paragraph)

        if options:
            data.append({
                'question': options[0],
                'options': options[1:],
                'correct_option': correct_option
            })

    return data

def create_answer_document(data, output_file):
    answer_doc = Document()

    for entry in data:
        question_text = entry['question']
        options_text = "\n".join(entry['options'])
        correct_option = entry['correct_option']

        if correct_option is not None:
            answer_text = f"{question_text}\n{options_text}\nANSWER: {correct_option}."
            print(f"Answer found: {answer_text}")

            answer_doc.add_paragraph(answer_text)

    print("\nFinal Output:")
    for paragraph in answer_doc.paragraphs:
        print(paragraph.text)

    answer_doc.save(output_file)

if __name__ == "__main__":
    input_file = r"C:\Users\Khizar Riyaz\Downloads\Product Knowledge - Liability Products (100 QUES).docx"
    output_file = r"C:\Users\Khizar Riyaz\Downloads\output.docx"

    data = process_input_file(input_file)
    create_answer_document(data, output_file)

    print(f"\nAnswer document created: {output_file}")
```
